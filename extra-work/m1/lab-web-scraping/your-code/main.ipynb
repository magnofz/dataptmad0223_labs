````python
df_2 = pd.DataFrame(datos, columns=['Lat','Lat d', 'Lon', 'Lon d', 'Region'])
````

#### Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'
````

````python
html_11 = requests.get(url, headers=headers).content
parsed_html_11 = bs4.BeautifulSoup(html_11, 'html.parser')
````

````python
tags_t = parsed_html_11.find_all('td', {'class': 'titleColumn'})
tags_r = parsed_html_11.find_all('td', {'class': 'ratingColumn imdbRating'})
````

````python
regex = '.+(?=\s\(dir)'

titles = [(tags_t[x].a.string,
           re.findall('\d+',tags_t[x].span.string)[0],
           re.findall(regex, tags_t[x].a['title'])[0],
           tags_r[x].strong.string) for x in range(len(tags_t))]
````

````python
columns = ['Movie Name', 'Initial Release', 'Director', 'Stars']
top_250 = pd.DataFrame(titles, columns=columns)
````

````python
top_250.head(5)
````
